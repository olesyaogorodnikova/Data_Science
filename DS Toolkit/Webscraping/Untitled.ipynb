```python
# Create the link (it is a better idea to constuct the especially link if you notice a specific pattern..)
link_first_part = 'https://www.atptour.com/' 
link_mid_1_part = 'en/players/roger-federer/f324/rankings-history'

url = link_first_part + link_mid_1_part
url
```

```python
# get the content from the link
response =requests.get(url, timeout=50)

# parse the content with BeautifulSoup with the html.parser method
soup = BeautifulSoup(response.content, "html.parser")

#check status code of the downloaded page
response.status_code
soup.findAll('tbody')
```

```python
columns = [
    'data',
    'ranking',

]
df = pd.DataFrame(columns = columns)
```

```python
for i in range(len(soup.find('tbody').findAll('tr'))):
    data=soup.find('tbody').findAll('tr')[i].findAll('td')[0].get_text().strip()
    ranking=soup.find('tbody').findAll('tr')[i].findAll('td')[1].get_text().strip()
   
    df = df.append({
    'data':data,
    'ranking':ranking        
    }, ignore_index=True)
```
