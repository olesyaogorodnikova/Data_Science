What is the smallest relative share of a Nobel prize ever given?

```python
data1 = data['prizes']
b=0
for prdct in data1:
    if 'laureates' in prdct:  
        laureates = prdct['laureates']
        for laureate in laureates:
            a = int(laureate['share'])
            if a>b:
                b=a
                
print(b)

```

```python
data1 = data['prizes']
dd = {}
for prdct in data1:
    category  = prdct['category']
    if 'laureates' in prdct:
        for person in prdct['laureates']:
            name = person['firstname'] + " "
            if 'surname' in person:
                name+= person['surname']
            if name not in dd:    
                dd[name] =  [category]                                               
            else:
                dd[name].append(category)
for name in dd:
    if len(dd[name])>1:
        print(name, dd[name])
```
